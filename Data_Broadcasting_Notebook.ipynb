# Received Data from Laptop

# Create a Pandas Data Frame

```python
import paho.mqtt.client as mqttsub
import pandas as pd

data=[]
count=0
client=mqttsub.Client()
client.connect('broker.hivemq.com',1883)
print('Broker Connected')
client.subscribe('iot/icibt')

def notification(client,userdata,msg):
  global count
  t=msg.payload
  t=t.decode('utf-8')
  t=t.split(' ')
  hum=float(t[1][1:-2])
  temp=float(t[-1][1:-2])
  label=0
  if(hum>20 and hum<40):
    label=1
  elif(hum>40 and hum<60):
    label=2
  elif(hum>60 and hum<80):
    label=3
  elif(hum>80 and hum<100):
    label=4
  dummy=[]
  dummy.append(hum)
  dummy.append(temp)
  dummy.append(label)
  data.append(dummy)
  
  count+=1
  if(count==10): # hum - col0, temp - col1
    df=pd.DataFrame(data)
    df.to_csv('iot.csv')
    count=0


  print(data)

client.on_message=notification
client.loop_forever()
```
